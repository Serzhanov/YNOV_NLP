```python
df2 = pd.read_table('data/villes.txt')
df2.head()
```

```python
df2.columns = df2.columns.str.strip()
df2['Nom Ville'] = df2['Nom Ville'] .apply(lambda x: x.strip())
df2['MAJ'] = df2['MAJ'] .apply(lambda x: x.strip())
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot using latitude and longitude
plt.figure(figsize=(10, 6))  # Adjust the size of the plot if needed
plt.scatter(df2['Longitude'][:10], df2['Latitude'][:10], marker='o', color='blue')

# Add city names as labels to the points
for i, row in df2[:10].iterrows():
    plt.annotate(row['Nom Ville'], (row['Longitude'], row['Latitude']), fontsize=8)

# Add titles and axis labels if needed
plt.title("Map of Cities")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.grid(True)
plt.show()
```

```python
import folium
import pandas as pd

# Create a Folium map centered on the average latitude and longitude
map_center = [df2['Latitude'][:10].astype(float).mean(), df2['Longitude'][:10].astype(float).mean()]
m = folium.Map(location=map_center, zoom_start=7)

# Add markers for each city
for i, row in df2[:10].iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], tooltip=row['Nom Ville']).add_to(m)

# Save the map as an HTML file or display it in a Jupyter Notebook
m.save('cities_map.html')
```
